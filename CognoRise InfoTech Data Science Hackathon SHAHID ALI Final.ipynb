# Handling missing values in Training dataset

```python
import pandas as pd

# Load the dataset
train_df = pd.read_csv('C&T train dataset1.csv')
  # Update with your dataset filename

# Display the number of missing values in each column
print("Number of missing values in each column:")
print(train_df.isnull().sum())


# Numeric Columns: Impute missing values with mean
numeric_columns = ['poi', 'age']
for col in numeric_columns:
    train_df[col].fillna(train_df[col].mean(), inplace=True)


    
    
    
    # Categorical Columns: Impute missing values with mode
categorical_columns = ['employment_st', 'housing_type','gurantors']
for col in categorical_columns:
    train_df[col].fillna(train_df[col].mode()[0], inplace=True)

    

# Display the number of missing values after handling
print("\nNumber of missing values after handling:")
print(train_df.isnull().sum())
```

# Split the datasets into features and target variable


```python
# Split the datasets into features and target variable
X_train = train_df.drop('Group_no', axis=1)
y_train = train_df['Group_no']
X_test = test_df.drop('credits_no', axis=1)
y_test = test_df['credits_no']
```

# Performing one-hot encoding on categorical variables in both training and testing datasets

```python
# Perform one-hot encoding on categorical variables in both training and testing datasets
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

# Ensure that both datasets have the same columns after one-hot encoding
# Add missing columns to the test dataset with default values of 0
#missing_cols = set(X_train.columns) - set(X_test.columns)
#for col in missing_cols:
#    X_test[col] = 0

# Reorder the columns in the test dataset to match the order in the train dataset
#X_test = X_test[X_train.columns]
```

# Split the data into training and testing sets


```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.25,random_state=42)

# Initialize classifiers
classifiers = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=0),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Logistic Regression": LogisticRegression(),
    "KNN": KNeighborsClassifier(),
    "NB": GaussianNB(),
    "MLP": MLPClassifier(),
    "Support Vector Machine": SVC(kernel='linear', random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0)
}
```

```python
X_train.shape
```

```python
X_test.shape
```

# Train and evaluate each classifier


```python
# Train and evaluate each classifier
for name, clf in classifiers.items():
    print(f"Training {name}...")
    # Train the classifier
    clf.fit(X_train, y_train)
    
    # Make predictions on the testing set
    y_pred = clf.predict(X_test)
    
    # Calculate accuracy
    print(" ")

    accuracy = accuracy_score(y_test, y_pred)
    print(f"                Accuracy for {name}: {accuracy*100:.2f}%")
    
    # Print classification report
    print(" ")

    print(f"Classification Report for {name}:")
    print(classification_report(y_test, y_pred))
    
    # Print confusion matrix
    print(f"Confusion Matrix for {name}:")
    print(confusion_matrix(y_test, y_pred))
    print("----------------------------------------------")
```
